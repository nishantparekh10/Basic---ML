```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,c='r',linestyle='dashed')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,a,'b')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Wave Form')
plt.plot(x,y)
```

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,8]
plt.bar(x,y, align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
```

```python
data = [np.random.normal(0,std,100) for std in range (1,4)]

plt.boxplot(data,vert=True,patch_artist=True);
```
